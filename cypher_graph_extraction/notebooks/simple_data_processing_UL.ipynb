```python
meta_file = '../hp-extractor/raw_data/hp_metadata.tsv'
abstracts_file = '../hp-extractor/raw_data/hp_abstracts.tsv'
names_file = '../hp-extractor/raw_data/abs_with_verbatim_names.csv'

meta = pd.read_csv(meta_file, sep='\t')
abstracts = pd.read_csv(abstracts_file, sep='\t')
names = pd.read_csv(names_file)
```

```python
names
```

```python
names_min = names[['absID', 'combine_abstract']]
```

```python
names_min
```

```python
names_min.to_csv('title_abstract_combined_with_names.csv', index=None)
```

```python
for index, row in abstracts.iterrows():
    
    absID = row['absID']

    title = meta[meta['absID'] == absID]['title'].values[0]
    abs = row['abs']
    
    if type(title) == float:
        title = ''

    combined_abstract = str(title) + ' ' + str(abs)

    abstracts.loc[index, 'title_abstract'] = combined_abstract
```

```python
abstracts_min = abstracts[['absID', 'title_abstract']]
```

```python
abstracts_min.to_csv('title_abstract_all.csv', index=None)
```
