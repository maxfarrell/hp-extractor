```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Initialize a directed graph
G = nx.DiGraph()

# Entities and their types
entities = {
    "Hantavirus": "Virus",
    "S genomic segment": "Genomic Segment",
    "New York (NY) virus": "Virus",
    "Four Corners (FC) virus": "Virus",
    "Sin Nombre virus": "Virus",
    "Hantavirus pulmonary syndrome (HPS)": "Disease",
    "White-footed mouse (Peromyscus leucopus)": "Host (Rodent)",
    "Deer mouse (Peromyscus maniculatus)": "Host (Rodent)",
    "Shelter Island, New York": "Location",
    "Nucleotide": "Genetic Material",
    "Amino acid": "Protein Component",
    "Phylogenetic relationship": "Relationship",
    "Maximum parsimony": "Analytical Method",
    "Neighbor-joining": "Analytical Method"
}

# Relationships
relationships = [
    ("White-footed mouse (Peromyscus leucopus)", "harbors", "Hantavirus"),
    ("Deer mouse (Peromyscus maniculatus)", "harbors", "Hantavirus"),
    ("White-footed mouse (Peromyscus leucopus)", "harbors", "New York (NY) virus"),
    ("Deer mouse (Peromyscus maniculatus)", "harbors", "Four Corners (FC) virus"),
    ("New York (NY) virus", "isolated from", "White-footed mouse (Peromyscus leucopus)"),
    ("New York (NY) virus", "captured on", "Shelter Island, New York"),
    ("New York (NY) virus", "molecularly linked to", "Fatal case of HPS in the northeastern United States"),
    ("New York (NY) virus", "more similar to", "Four Corners (FC) virus"),
    ("New York (NY) virus", "distinct from", "Other sigmodontine rodent-borne hantaviruses"),
    ("S genomic segment of NY virus", "length", "2078 nucleotides"),
    ("Open reading frame", "length", "1284 nucleotides"),
    ("Protein encoded", "length", "428 amino acids"),
    ("3'-noncoding region", "length", "752 nucleotides"),
    ("New York (NY) virus", "differs from FC virus at nucleotide level by", "16.6-17.8%"),
    ("New York (NY) virus", "differs from FC virus at amino acid level by", "7.0-8.2%"),
    ("New York (NY) virus", "formed a separate lineage from", "Four Corners (FC) virus"),
    ("New York (NY) virus and FC virus", "relationship", "Phylogenetically distinct"),
    ("Phylogenetic analysis methods", "used for analysis", "Maximum parsimony"),
    ("Phylogenetic analysis methods", "used for analysis", "Neighbor-joining")
]

# Add nodes with types
for entity, entity_type in entities.items():
    G.add_node(entity, entity_type=entity_type)

# Add edges with relationships
for src, rel, dst in relationships:
    G.add_edge(src, dst, relationship=rel)

# Define node colors based on their types
color_map = {
    "Virus": "lightblue",
    "Genomic Segment": "lightgreen",
    "Disease": "lightcoral",
    "Host (Rodent)": "orange",
    "Location": "lightpink",
    "Genetic Material": "lightyellow",
    "Protein Component": "lightgrey",
    "Relationship": "lightpurple",
    "Analytical Method": "lightcyan"
}
#node_colors = [color_map[G.nodes[node]['entity_type']] for node in G]

# Define edge labels
edge_labels = {(src, dst): data['relationship'] for src, dst, data in G.edges(data=True)}

# Plotting the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)  # positions for all nodes
#nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=3000, font_size=10, font_weight="bold", edge_color="grey")
nx.draw(G, pos, with_labels=True, node_size=3000, font_size=10, font_weight="bold", edge_color="grey")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', font_size=8)
plt.title('Knowledge Graph of Hantavirus Entities and Relationships')
plt.show()
```
